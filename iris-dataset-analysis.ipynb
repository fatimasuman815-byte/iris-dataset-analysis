```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## ðŸ“‚ Loading the Dataset
We load the Iris dataset using seaborn.

```python
iris = sns.load_dataset("iris")
```

## ðŸ“Š Scatter Plot
This plot shows the relationship between petal length and petal width.


```python
sns.scatterplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    hue="species"
)
plt.show()
```

## ðŸ“Š Histogram
This plot shows the distribution of sepal length values.

```python
sns.histplot(iris["sepal_length"], bins=20, kde=True)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

### ðŸ” Observation (Histogram)
Sepal length follows an approximately normal distribution,  
and no extreme outliers are observed.

## ðŸ“¦ Box Plot
This box plot compares petal length across different species.

```python
sns.boxplot(
    data=iris,
    x="species",
    y="petal_length"
)
plt.title("Petal Length by Species")
plt.show()
```

### ðŸ” Observation (Box Plot)
- *Setosa* has the smallest petal length values  
- *Virginica* has the largest petal length values  
- No extreme outliers are present in the dataset
